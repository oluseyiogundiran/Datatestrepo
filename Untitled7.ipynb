```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# read dataframe
df = pd.read_csv("adult.csv")
```

```python
# view df
df.head()
```

```python
# replace null values
df.replace('?', np.nan, inplace = True)
df.isnull().sum()

# check extra spaces
```

```python
# Question 1: How many men and women (sex feature) are represented in this dataset?
df["gender"].value_counts()
```

```python
# Question 2: What is the average age (age feature) of women?
df.loc[df['gender'].str.strip() == 'Female', 'age'].mean()
```

```python
# Question 3: What is the percentage of German citizens (native-country feature)?
german_citizens = (df['native-country'].str.strip() == 'Germany').mean() * 100
"{:.2f}%".format(german_citizens)
```

```python
(df[df['native-country']=='Germany'].count() /df['native-country'].value_counts().sum())['native-country']
```

```python
# Question 4: What are the mean and standard deviation of age for those who earn more than
# 50K per year (salary feature) and those who earn less than 50K per year?

ages1 = df.loc[df['income'].str.strip() == '>50K', 'age']
ages1_mean = round(ages1.mean())
ages1_std = round(ages1.std(), 1)

ages2 = df.loc[df['income'].str.strip() == '<=50K', 'age']
ages2_mean = round(ages2.mean())
ages2_std = round(ages2.std(), 1)

result = "The average age of the rich: {0} +- {1} years, poor - {2} +- {3} years.".format(
    ages1_mean, ages1_std, ages2_mean, ages2_std
)

result
```

```python
# Question 7: Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)?
# Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.
df['marital-status'].unique()
mapping={'Never-married':'Bachelor', 'Divorced':'Bachelor','Separated':'Bachelor','Widowed':'Bachelor','Married-civ-spouse':'Married',
        'Married-spouse-absent':'Married','Married-AF-spouse':'Married'}

df['status'] = df['marital-status'].map(mapping)

df[df['income'] == '>50K']['status'].value_counts()
```
