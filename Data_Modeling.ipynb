```python
bin_edges = [0, 100, 200, 300, 400]
bin_labels = ['0-100', '101-200', '201-300', '301-400']
df['Recovered_Bins'] = pd.cut(df['Recovered'], bins=bin_edges, labels=bin_labels)
df.head()
```
