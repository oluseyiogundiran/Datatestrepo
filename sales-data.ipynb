Exercise 2: Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

```python
plt.figure(figsize=(10, 6))
plt.plot(sales_data['month_number'], sales_data['total_profit'], linestyle=':', color='red',
         marker='o', markerfacecolor='red', linewidth=3, label='Total Profit')
plt.title('Monthly Total Profit (Styled)')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

Exercise 3: Read all product sales data and show it using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline
for each product)

```python
# Exercise 3: Read all product sales data and show it using a multiline plot

plt.figure(figsize=(12, 8))

# Plotting each product sales data
plt.plot(sales_data['month_number'], sales_data['facecream'], label='Face Cream', marker='o')
plt.plot(sales_data['month_number'], sales_data['facewash'], label='Face Wash', marker='o')
plt.plot(sales_data['month_number'], sales_data['toothpaste'], label='Toothpaste', marker='o')
plt.plot(sales_data['month_number'], sales_data['bathingsoap'], label='Bathing Soap', marker='o')
plt.plot(sales_data['month_number'], sales_data['shampoo'], label='Shampoo', marker='o')
plt.plot(sales_data['month_number'], sales_data['moisturizer'], label='Moisturizer', marker='o')

plt.title('Monthly Sales Data of Products')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend()
plt.grid(True)
plt.show()
```

Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot. Also, add a grid in
the plot. gridline style should 

```python
# Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot. Also, add a grid in the plot. gridline style should .

plt.figure(figsize=(10, 6))
plt.scatter(sales_data['month_number'], sales_data['toothpaste'], color='blue')
plt.title('Monthly Toothpaste Sales')
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Units Sold')
plt.grid(True, linestyle='--')
plt.show()
```

```python
# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

plt.figure(figsize=(12, 8))
bar_width = 0.35
index = sales_data['month_number']

# Creating bars for face cream and facewash
plt.bar(index, sales_data['facecream'], bar_width, label='Face Cream')
plt.bar(index + bar_width, sales_data['facewash'], bar_width, label='Face Wash')

plt.title('Monthly Sales of Face Cream and Facewash')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.xticks(index + bar_width / 2, index)
plt.legend()
plt.show()
```

```python
# Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

plt.figure(figsize=(10, 6))
plt.bar(sales_data['month_number'], sales_data['bathingsoap'], color='blue')
plt.xlabel('Month Number')
plt.ylabel('Sales of Bathing Soap')
plt.title('Monthly Sales of Bathing Soap')
plt.xticks(sales_data['month_number'])
plt.grid(True, linestyle='--', alpha=0.5)

# Save the plot
bathing_soap_sales_chart_path = 'bathing_soap_sales_chart.png'
plt.savefig(bathing_soap_sales_chart_path)

bathing_soap_sales_chart_path
```

```python
# Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

plt.figure(figsize=(10, 6))
plt.hist(sales_data['total_profit'], bins=10, color='green', alpha=0.7)
plt.xlabel('Profit Range')
plt.ylabel('Frequency')
plt.title('Distribution of Total Monthly Profits')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

```python
product_totals = sales_data[['facecream','facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()

plt.figure(figsize=(8,8))
plt.pie(product_totals, labels=product_totals.index)
plt.title('Total Sale Data For Last Year For Each Product')
plt.show()
```

```python
# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

fig, axs = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))

# Bathing Soap Sales
axs[0].plot(sales_data['month_number'], sales_data['bathingsoap'], marker='o', color='blue')
axs[0].set_title('Monthly Sales of Bathing Soap')
axs[0].set_xlabel('Month Number')
axs[0].set_ylabel('Units Sold')

# Facewash Sales
axs[1].plot(sales_data['month_number'], sales_data['facewash'], marker='o', color='orange')
axs[1].set_title('Monthly Sales of Facewash')
axs[1].set_xlabel('Month Number')
axs[1].set_ylabel('Units Sold')

plt.tight_layout()
plt.show()
```

```python
# Exercise 10: Read all product sales data and show it using the stack plot

plt.figure(figsize=(12, 8))

# Creating a stack plot for all products
plt.stackplot(sales_data['month_number'],
              sales_data['facecream'],
              sales_data['facewash'],
              sales_data['toothpaste'],
              sales_data['bathingsoap'],
              sales_data['shampoo'],
              sales_data['moisturizer'],
              labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])

plt.title('Monthly Sales Data of Products (Stack Plot)')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.legend(loc='upper left')
plt.show()
```
