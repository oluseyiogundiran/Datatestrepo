```python
import pandas as pd
import matplotlib.pyplot as plt


path = 'https://courses.washington.edu/b517/Datasets/SalaryData.csv'
df= pd.read_csv(path)
df.head()
```

**Background story of the dataset**:

This DataSet examines salary disparities at a US University/College, focusing on potential sex bias in faculty salaries. This is a significant issue given that salary discrimination based on sex is illegal in the US. The dataset aims to analyze whether differences in salaries between male and female faculty members are due to sex bias or other factors like experience, degree, field of work, administrative responsibilities, and productivity.



```python
average_salary_by_year_sex = df.groupby(['year', 'gender'])['salary'].mean().unstack()

# Plotting the bar chart
average_salary_by_year_sex.plot(kind='bar', figsize=(12, 6), colormap='viridis')
plt.title('Average Salaries by Sex Over the Years')
plt.xlabel('Years')
plt.ylabel('Average Salary')
plt.legend(title='Sex', loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Plot Use Case:**

The bar chart visually represents the average salaries of male and female faculty members over the years. Each bar represents a specific year, and for each year, there are two bars (one for male and one for female), allowing for a direct comparison of average salaries between the two sexes.
The color differentiation (colormap) helps distinguish between male and female bars.
We can observe trends, variations, or potential biases in average salaries over the years, providing a clear visual representation of the dataset's information.


**Insight from the Bar Chart:**


1. **Trend Over Time**: Both male (M) and female (F) average salaries have generally increased over the years.

2. **Gender Pay Gap**: There is a visible gap between the average salaries of males and females, with males consistently earning more on average than females in each year presented.

3. **Consistency of Gap**: The gap between male and female salaries appears to be fairly consistent across the years. This suggests that whatever factors contribute to this gap have remained relatively stable over this time period.





**Rationale for choosing this particular visualization:**

The choice of a bar chart in this instance seems appropriate for the purpose of comparing two categories across a third variable(s). In this case, third variables are the years, and within each year, there are two categories: male and female.

Also, this Bar chart allows for a clear visual distinction between genders, with different colors making it easy to differentiate between male and female average salaries.
