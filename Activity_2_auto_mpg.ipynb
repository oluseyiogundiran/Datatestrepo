```python
import pandas as pd
```

```python
df = pd.read_excel('auto-mpg.xlsx')
df.head()
```

```python
#Drop column Origin from the dataset
df.drop(['origin'], axis=1,inplace=True)
df.head()
```

```python
# change columns' name with a space
df.columns = df.columns.str.replace(' ', '_')
df.head()
```

```python
#Find out all the missing values in the dataset and replace it with its
#most appropriate replacement
df.isnull().sum()
```

```python
import seaborn as sns
# Box plot for acceleration
sns.boxplot(df.acceleration)
```

```python
# Distribution plot for acceleration
sns.distplot(df.acceleration)
```

```python
# Repeat for model_year
sns.boxplot(df.model_year)
```

```python
sns.distplot(df.model_year)
```

```python
df_imput = df.copy()
# replace null in 'acceleration' with MODE
df_imput['acceleration'] = df_imput['acceleration'].fillna(df_imput['acceleration'].mode()[0])
#check imputation success
#print (df_imput.isnull().sum())

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.distplot(df_imput['acceleration'], label='Imputed')
sns.distplot(df['acceleration'], label='Original')
plt.legend()
plt.show()
```

```python
# replace null in 'model year' with MODE
df_imput['model_year'] = df_imput['model_year'].fillna(df_imput['model_year'].mode()[0])
#sns.distplot(df_imput.model_year)
#check imputation success
print (df_imput.isnull().sum())

plt.figure(figsize=(10, 6))
sns.distplot(df_imput['model_year'], label='Imputed')
sns.distplot(df['model_year'], label='Original')
plt.legend()
plt.show()
```

```python
# replace all void 'CAR_Number' with a string 'noCarNum'
df_imput['CAR_Number'] = df_imput['CAR_Number'].fillna('noCarNum')
# check a success
print (df_imput.isnull().sum())
```

```python
df_imput.sort_values('CAR_Number', ascending=False).head(50)
```

```python
# remove all duplicate entries (rows) for the column  CAR Number
df_imput = df_imput.drop_duplicates(subset='CAR_Number')
df_imput.sort_values('CAR_Number', ascending=False).head(50)
```

```python
# Display the first five and last five rows
print (df_imput.tail(5))
df_imput.head(5)
```

```python
# Display all the column names in the dataset
df_imput.columns
```

```python
# Display the concise summary of your dataset
df_imput.info()
```

```python
# Display the name of the car with maximum number of horsepower
max_horsepower_car = df_imput['car_name'][df_imput['horsepower'].idxmax()]
print(max_horsepower_car, df_imput['horsepower'].max(), 'hp')
```

```python
# change the name of the column "mpg" to "L/100km"
df_imput = df_imput.rename(columns={'mpg': 'L/100km'})
df_imput.head()
```

```python
# change values in 'L/100km'
df_imput['L/100km'] = round((235.215 / df_imput['L/100km']),2)
df_imput.head()
```

```python
# Normalize the column Weight so that the values range from 0 to 1 (original value)/(maximum value)
df_imput['weight'] = (df_imput['weight'] / (df_imput['weight'].max()))
df_imput.head()
```

```python
#  rearrange horsepower into three bins' to simplify analysis
df_imput['horsepower_bin'] = pd.cut(df_imput['horsepower'], bins=3, labels=['Low', 'Medium', 'High'])
df_imput.head()
```

```python
# a histogram to visualize the distribution of bins we created above
df_imput['horsepower_bin'].value_counts().plot(kind='bar')
plt.title('Horsepower Bins')
plt.xlabel('Bin')
plt.ylabel('Count')
plt.show()
```
